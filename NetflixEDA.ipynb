```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (20,10)
```

```python
df = pd.read_csv("netflix_titles.csv")
df[['Clas1','Clas2','Clas3']]=df['listed_in'].str.split(',', expand=True)
df['date_added']= pd.to_datetime(df['date_added'])
#df['cast'].str.split(',', expand=True)
```

```python
top10 = ['United States','India','United Kingdom','Japan','South Korea','Canada','Spain','France','Mexico','Egypt']
```

```python
df_usa = df[(df['country']=="United States") & (df['date_added']>"2014-01-01")]
df_spa = df[(df['country']=="Spain")]
df_ind = df[df['country']=="India"]
df_uk = df[df['country']=="United Kingdom"]
df_jap = df[df['country']=="Japan"]
df_cor = df[df['country']=="South Korea"]
df_can = df[df['country']=="Canada"]
df_fr = df[df['country']=="France"]
df_mex = df[df['country']=="Mexico"]
df_egy = df[df['country']=="Egypt"]
```

Producciones USA 2014-2022

```python
df_usa.groupby([pd.Grouper(key='date_added', freq='M'), 'type']).size().unstack().plot(kind='line')
```

```python
df_spa.groupby([pd.Grouper(key='date_added', freq='M'), 'type']).size().unstack().plot(kind='line')
```

Producciones India 2014-2022

```python
df_ind.groupby([pd.Grouper(key='date_added', freq='M'), 'type']).size().unstack().plot(kind='line')
```

```python
df_usa_month=df_usa.groupby([pd.Grouper(key='date_added', freq='M'), 'type']).size().unstack()
df_spa_month=df_spa.groupby([pd.Grouper(key='date_added', freq='M'), 'type']).size().unstack()
df_ind_month=df_ind.groupby([pd.Grouper(key='date_added', freq='M'), 'type']).size().unstack()
df_uk_month=df_uk.groupby([pd.Grouper(key='date_added', freq='M'), 'type']).size().unstack()
df_jap_month=df_can.groupby([pd.Grouper(key='date_added', freq='M'), 'type']).size().unstack()
df_can_month=df_uk.groupby([pd.Grouper(key='date_added', freq='M'), 'type']).size().unstack()
df_cor_month=df_cor.groupby([pd.Grouper(key='date_added', freq='M'), 'type']).size().unstack()
df_fr_month=df_fr.groupby([pd.Grouper(key='date_added', freq='M'), 'type']).size().unstack()
df_mex_month=df_mex.groupby([pd.Grouper(key='date_added', freq='M'), 'type']).size().unstack()
df_egy_month=df_egy.groupby([pd.Grouper(key='date_added', freq='M'), 'type']).size().unstack()
```

```python
with plt.style.context('Solarize_Light2'):
    plt.plot(df_usa_month.index, df_usa_month['Movie'], label="USA")
    plt.plot(df_spa_month.index, df_spa_month['Movie'], label = "SPAIN")
    plt.plot(df_ind_month.index, df_ind_month['Movie'],label = "INDIA")
    plt.plot(df_uk_month.index, df_uk_month['Movie'],label = "UK")
    plt.plot(df_jap_month.index, df_jap_month['Movie'],label = "JAPAN")
    plt.plot(df_cor_month.index, df_cor_month['Movie'],label = "KOREA")
    plt.plot(df_can_month.index, df_can_month['Movie'],label = "CANADA")
    plt.plot(df_fr_month.index, df_fr_month['Movie'],label = "FRANCE")
    plt.plot(df_mex_month.index, df_mex_month['Movie'],label = "MEXICO")
    plt.plot(df_egy_month.index, df_egy_month['Movie'],label = "EGYPT")

    # Number of accent colors in the color scheme
    plt.title('Peliculas producidas por país y año')
    plt.xlabel('Año', fontsize=14)
    plt.ylabel('Número de producciones', fontsize=14)
plt.legend()
plt.show()
```

```python
with plt.style.context('Solarize_Light2'):
    plt.plot(df_usa_month.index, df_usa_month['TV Show'], label="USA")
    plt.plot(df_spa_month.index, df_spa_month['TV Show'], label = "SPAIN")
    plt.plot(df_ind_month.index, df_ind_month['TV Show'],label = "INDIA")
    plt.plot(df_uk_month.index, df_uk_month['TV Show'],label = "UK")
    plt.plot(df_jap_month.index, df_jap_month['TV Show'],label = "JAPAN")
    plt.plot(df_cor_month.index, df_cor_month['TV Show'],label = "KOREA")
    plt.plot(df_can_month.index, df_can_month['TV Show'],label = "CANADA")
    plt.plot(df_fr_month.index, df_fr_month['TV Show'],label = "FRANCE")
    plt.plot(df_mex_month.index, df_mex_month['TV Show'],label = "MEXICO")
    plt.plot(df_egy_month.index, df_egy_month['TV Show'],label = "EGYPT")

    # Number of accent colors in the color scheme
    plt.title('Series producidas por país y año')
    plt.xlabel('Año', fontsize=14)
    plt.ylabel('Número de producciones', fontsize=14)
plt.legend()
plt.show()
```

Producciones por meses en USA y España

```python
df_usa_month['date']=df_usa_month.index
df_usa_month['date'] = df_usa_month['date'].astype(str)
df_usa_month [['year', 'month', 'day']] = df_usa_month['date'].str.split('-', expand=True)

df_spa_month['date']=df_spa_month.index
df_spa_month['date'] = df_spa_month['date'].astype(str)
df_spa_month [['year', 'month', 'day']] = df_spa_month['date'].str.split('-', expand=True)
```

```python
x = ['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
default_x_ticks = range(len(x))

df_usa_month.groupby('month').sum().plot(kind='bar')
plt.xticks(default_x_ticks, x)
df_spa_month.groupby('month').sum().plot(kind='bar')

plt.xticks(default_x_ticks, x)
```
